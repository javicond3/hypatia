```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Colores para las columnas
column_colors = {
    'Mean RTT_telesat': '#1f77b4',
    'Mean RTT_kuiper': '#ff7f0e',
    'Mean RTT_starlink': '#2ca02c',
    'avg_of': '#d62728'
}

column_names = {
    'Mean RTT_telesat': 'Telesat',
    'Mean RTT_kuiper': 'Kuiper',
    'Mean RTT_starlink': 'Starlink',
    'avg_of': 'Optical fiber'
}
```

```python
df = pd.read_csv("merged_of_satellite_satellite_number_clean.csv")
df = df.dropna(subset=['Distance_of'])
cities = pd.concat([df['origin_name'], df['destination_name']]).unique()
cities
```

```python
for city in cities:
    print(city)
    city_connections = df[(df['origin_name'] == city) | (df['destination_name'] == city)]
    data = city_connections.copy()
    data.loc[:, 'Distance_of'] = data['Distance_of'].str.replace(',', '.').astype(float)
    data.loc[:, 'avg_of'] = data['avg_of'].str.replace(',', '.').astype(float)
    data = data.sort_values(by='Distance_of') 

    plt.scatter(data['Distance_of'], data['Mean RTT_telesat'], label='Telesat', s=10)
    plt.scatter(data['Distance_of'], data['Mean RTT_kuiper'], label='Kuiper', s=10)
    plt.scatter(data['Distance_of'], data['Mean RTT_starlink'], label='Starlink', s=10)
    plt.scatter(data['Distance_of'], data['avg_of'], label='Optical Fiber', s=10)


    y_max_pos = data[['Mean RTT_telesat', 'Mean RTT_kuiper', 'Mean RTT_starlink', 'avg_of']].max().max()*1.3
    i = 0
    for index, row in data.iterrows():
        i += 1
        if row['origin_name'] == city:
            city_name = row['destination_name']
        else:
            city_name = row['origin_name']
        
        if i % 2 == 0:
            penalty = + 10
        else:
            penalty = -10
        plt.text(row['Distance_of'], y_max_pos + penalty, city_name, ha='center', va='top', rotation=70, fontsize=8)

    plt.title(f"{city}")
    plt.xlabel('Distance (km)')
    plt.ylabel('Mean RTT (ms)')
    plt.ylim(0, y_max_pos*1.3)
    plt.legend(loc='upper left', ncol=4)
    

    plt.show()
```

```python
data = df.copy()
data.loc[:, 'Distance_of'] = data['Distance_of'].str.replace(',', '.').astype(float)
data.loc[:, 'avg_of'] = data['avg_of'].str.replace(',', '.').astype(float)
data.loc[:, 'Mean RTT_telesat'] = data['Mean RTT_telesat'].astype(float)
data.loc[:, 'Mean RTT_kuiper'] = data['Mean RTT_kuiper'].astype(float)
data.loc[:, 'Mean RTT_starlink'] = data['Mean RTT_starlink'].astype(float)

data = data.sort_values(by='Distance_of') 
distance_step = 2000

data['winner'] = data[['Mean RTT_telesat', 'Mean RTT_kuiper', 'Mean RTT_starlink', 'avg_of']].idxmin(axis=1)
data['distance_group'] = (data['Distance_of'] // distance_step * distance_step).astype(int)

categories = ['Mean RTT_telesat', 'Mean RTT_kuiper', 'Mean RTT_starlink', 'avg_of']
num_categories = len(categories)

# Agrupar los datos por distancia y tipo de ganador
grouped_data = data.groupby(['distance_group', 'winner']).size().unstack().sort_index()


# Crear el grfico de barras
ax = grouped_data.plot(kind='bar', stacked=True, color=[column_colors.get(col, '#1f77b4') for col in grouped_data.columns])

legend_labels = [column_names.get(col, col) for col in grouped_data.columns]
ax.legend(legend_labels, title='')



def getxTicks(x_group_points, distance_step):
    new_x = []
    for x in x_group_points:
        x = int(x)
        new_x.append(str(x // distance_step * distance_step) + "-" + str(x // distance_step * distance_step + distance_step))
    return new_x


# Establecer etiquetas y ttulo
plt.xlabel('Distance (km)')
plt.ylabel('Number of occurrences')
plt.title('Lowest RTT technology')

x_group_points = data['distance_group'].unique()
x_ticks = getxTicks(x_group_points, distance_step)  # Obtener las etiquetas correspondientes a las ubicaciones

plt.xticks(range(len(x_group_points)), x_ticks)  # Establecer las etiquetas del eje x

# Mostrar el grfico
plt.show()
```
